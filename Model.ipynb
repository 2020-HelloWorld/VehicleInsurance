```python
from sklearn.metrics.pairwise import cosine_similarity
DF_ORIGIN=pd.read_csv('df.csv')
DF_ORIGIN.dropna()
def find_car(df_make,df_model,df_variant,df_fuel,df_localcode):

    def make_classify(master_make,df_make,master_make_unique):
        make_embed = model.encode([df_make])
        make_embed=make_embed.reshape(1, -1)
        cs=cosine_similarity(
        make_embed,
        master_make[:]
        )
        di=list(zip(master_make,cs[0],master_make_unique))
        pp=(sorted(list(di),key=lambda x: x[1],reverse=True))
        car_make=pp[0][2]
        df=master.copy()
        return df[df['make']==car_make]
        
        
    extracted_makedf=make_classify(master_make,df_make,master_make_unique)

    master_model_unique=list(extracted_makedf.model.unique())
    master_model = model.encode(master_model_unique)
    master_model.shape



    def model_classify(master_model,df_model,master_model_unique):
        make_embed = model.encode([str(df_model)])
        make_embed=make_embed.reshape(1, -1)
        cs=cosine_similarity(
        make_embed,
        master_model[:]
        )
        di=list(zip(master_model,cs[0],master_model_unique))
        pp=(sorted(list(di),key=lambda x: x[1],reverse=True))
        car_model=pp[0][2]
        df=extracted_makedf.copy()
        return df[df['model']==car_model]

    extracted_modeldf=model_classify(master_model,df_model,master_model_unique)

    master_variant_unique=list(extracted_modeldf.variant.unique())
    master_variant = model.encode(master_variant_unique)
    master_variant.shape

    def variant_classify(master_variant,df_variant,master_variant_unique):
        make_embed = model.encode([str(df_variant)])
        make_embed=make_embed.reshape(1, -1)
        cs=cosine_similarity(
        make_embed,
        master_variant[:]
        )
        di=list(zip(master_variant,cs[0],master_variant_unique))
        pp=(sorted(list(di),key=lambda x: x[1],reverse=True))
        car_variant=pp[0][2]
        df=extracted_modeldf.copy()
        return df[df['variant']==car_variant]

    extracted_variantdf=variant_classify(master_variant,df_variant,master_variant_unique)

    master_fuel_unique=list(extracted_variantdf.fuel_type.unique())
    master_fuel = model.encode(master_fuel_unique)


    def fuel_classify(master_fuel,df_fuel,master_fuel_unique):
        make_embed = model.encode([str(df_fuel)])
        make_embed=make_embed.reshape(1, -1)
        cs=cosine_similarity(
        make_embed,
        master_fuel[:]
        )
        di=list(zip(master_fuel,cs[0],master_fuel_unique))
        pp=(sorted(list(di),key=lambda x: x[1],reverse=True))
        car_model=pp[0][2]
        df=extracted_variantdf.copy()
        return df[df['fuel_type']==car_model]

    extracted_fueldf=fuel_classify(master_fuel,df_fuel,master_fuel_unique)
    return (int(extracted_fueldf.output), df_localcode)
for i in range(len(DF_ORIGIN)):
    try:
        print(find_car(DF_ORIGIN.make[i],DF_ORIGIN.model[i],DF_ORIGIN.variant,DF_ORIGIN.fuel_type[i],DF_ORIGIN.local_code[i]))
    except:
        print(DF_ORIGIN.local_code[i],'NONE')


```
